## Sample from random distribution:

Reminding myself how to get samples of a random distribution using Jacobians:
If we want a uniform distribution contained in a circle of radio 1, then find the 
cumulative distribution using polar coordinates, and invert it as per the usual trick:

```python
import random
import math
```

```python
#Test with 10000 points
samples = []
for _ in range(10000):
    u = math.sqrt(random.random())
    theta = 2*math.pi*random.random()
    samples.append((u*math.cos(theta), u*math.sin(theta)))
```

```python
samples[:3]
```

```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
#Plotting the distribution
fig = plt.figure(figsize=(10,10))
plt.scatter([x[0] for x in samples], [x[1] for x in samples])
plt.show()
```
